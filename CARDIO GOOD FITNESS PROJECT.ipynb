## Data Dictionary
- The data is about customers of the treadmill product(s) of a retail store called Cardio Good Fitness. It contains the following variables-

- Product - The model no. of the treadmill
- Age -  Age of the customer in no of years
- Gender - Gender of the customer
- Education - Education of the customer in no. of years
- Marital Status - Marital status of the customer
- Usage - Avg. # times the customer wants to use the treadmill every week
- Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit)
- Income - Income of the customer
- Miles- Miles that a customer expects to run

### MULTIVARIATE ANALYSIS

```python
# Create a correlation matrix to identify relationship between numerical variables
corr_mar = df.corr()
sns.heatmap(corr_mar, annot = True , cmap= "coolwarm")
plt.title("Correlation Matrix");
```

```python
# Pairplot to visualize relationships between numeric variables
sns.pairplot(df, hue='Product')
plt.suptitle('Pairplot of Numeric Variables by Product', y=1.02)
plt.show()
```

### Insights:

- Income Distribution:
TM798 customers generally have higher incomes compared to TM195 and TM498.
Insight: TM798 may be positioned as a premium product, appealing to customers with higher income levels.

- Product Usage:
TM195 and TM498 have a similar usage pattern, with most customers using the product 3-4 times a week.
TM798 users tend to have a higher usage frequency, possibly indicating a more dedicated and fitness-conscious customer base.
Insight: TM798 may attract customers with a more serious commitment to fitness.

- Fitness Levels:
TM798 customers generally have higher reported fitness levels compared to TM195 and TM498.
Insight: TM798 may be perceived as a product for customers with higher fitness goals.

- Correlations:
Positive correlations between income and product usage suggest that customers with higher incomes might use the products more frequently.
Positive correlations between age and income could indicate that older customers tend to have higher incomes.
